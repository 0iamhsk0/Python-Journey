- Simulate a restaurant order system.
- Menu has items with prices ()
- Customer orders multiple items ()
Tasks:


1. Show menu to the user.
2. Let them choose 5 items (with quantity), and store the order.
3. Calculate total bill.

Create a of unique ordered items.
Convert the final order summary to a for record-keeping.
: Update the menu dynamically (e.g., add a new item or change price).


```python
# Restaurant Menu & Orders Simulation

# Task 1: Show the Menu
menu = {
    "Starters": {
        101: {"Item": "Veg Spring Rolls", "Price": 120},
        102: {"Item": "Chicken Wings", "Price": 180},
        103: {"Item": "Paneer Tikka", "Price": 160}
    },
    "Main Course": {
        201: {"Item": "Butter Chicken", "Price": 260},
        202: {"Item": "Paneer Butter Masala", "Price": 220},
        203: {"Item": "Veg Biryani", "Price": 200},
        204: {"Item": "Chicken Biryani", "Price": 240}
    },
    "Desserts": {
        301: {"Item": "Gulab Jamun", "Price": 90},
        302: {"Item": "Ice Cream Sundae", "Price": 110}
    },
    "Beverages": {
        401: {"Item": "Lemon Soda", "Price": 60},
        402: {"Item": "Cold Coffee", "Price": 100},
        403: {"Item": "Masala Chai", "Price": 50}
    }
}

# Display menu
print("===== Restaurant Menu =====")
for category, items in menu.items():
    print(f"\n-- {category} --")
    for code, details in items.items():
        print(f"{code}: {details['Item']} - ₹{details['Price']}")

# Task 2: Take input for 5 items with quantity
print("\n===== Place Your Order =====")
order = {}
for i in range(5):
    code = int(input(f"Enter item code #{i+1}: "))
    qty = int(input("Enter quantity: "))
    order[code] = order.get(code, 0) + qty

# Task 3: Calculate Total Bill
print("\n===== Order Summary =====")
total = 0
unique_items = set()
for category in menu.values():
    for code, details in category.items():
        if code in order:
            item_total = order[code] * details['Price']
            total += item_total
            unique_items.add(details['Item'])
            print(f"{details['Item']} x {order[code]} = ₹{item_total}")

print(f"\nTotal Bill: ₹{total}")

# Task 4: Set of unique items ordered
print("\nUnique Items Ordered:")
print(unique_items)

# Task 5: Convert order summary to JSON for record-keeping
import json

order_summary = {
    "Ordered Items": [
        {"Item": item, "Quantity": order[code]}
        for category in menu.values()
        for code, details in category.items()
        if code in order and (item := details['Item'])
    ],
    "Total Amount": total
}

with open("order_record.json", "w") as file:
    json.dump(order_summary, file, indent=4)

# Task 6: Update menu dynamically
# Example: Adding a new item
menu["Desserts"][303] = {"Item": "Brownie", "Price": 130}
# Example: Changing price of an item
menu["Main Course"][201]["Price"] = 275

print("\nUpdated Menu Item: Brownie and New Price of Butter Chicken:")
print(menu["Desserts"][303])
print(menu["Main Course"][201])
```
