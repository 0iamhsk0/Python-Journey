### THE BINARY SEARCH
-> time complexity = O(log N)

-> note that binary_search() takes a condition as an argument

```python
def binary_search(low, high, cond):
    while low <= high:
        mid = (low + high)//2
        result = cond(mid)
        if result == 'found':
            return mid
        elif result == 'left':
            high = mid - 1
        elif result == 'right':
            high = mid + 1
    return -1
```

```python
def locate_card(cards, query) :
    #function closure
    def condition(mid):
        if cards[mid] == query:
            if mid > 0 and cards[mid-1] == query:
                return 'left'
            else:
                return 'found'
        elif cards[mid] < query:
            return 'left'
        else:
            return 'right'

    return binary_search(0, len(cards) - 1, condition)
```
