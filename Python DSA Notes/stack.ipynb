```python
class Stack:
    def __init__(self, size):
        self.size = size
        self.stack = []
        self.top = -1

    def push(self, val):
        if self.top == self.size - 1:
            print("Stack is full (Overflow condition)")
        else:
            self.top += 1
            self.stack.append(val)
            print(f"{val} has been pushed")

    def pop(self):
        if self.top == -1:
            print("Stack is empty (Underflow condition)")
        else:
            popped_value = self.stack.pop()
            self.top -= 1
            print(f"Popped value: {popped_value}")

    def find_top(self):
        if self.top == -1:
            print("No element in stack")
        else:
            print(f"Top element: {self.stack[self.top]}")

    def display_stack(self):
        if self.top == -1:
            print("Stack is empty")
        else:
            print("Stack contents (top to bottom):", self.stack[::-1])

# Get stack size from user
size = int(input("Enter the size of the stack: "))
stack = Stack(size)

# Pushing elements
print("Enter numbers to push into the stack (enter 'none' to stop):")
while True:
    user_input = input()
    if user_input.lower() == "none":
        break
    else:
        try:
            value = int(user_input)
            stack.push(value)
        except ValueError:
            print("Invalid input, please enter an integer")

# Display stack after input is done
print("\nFinal Stack Contents:")
stack.display_stack()

# Performing operations
while True:
    print("\nChoose an operation:\n1. Push\n2. Pop\n3. Find Top\n4. Display Stack\n5. Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        val = input("Enter a value to push: ")
        try:
            stack.push(int(val))
        except ValueError:
            print("Invalid input, please enter an integer")
    elif choice == "2":
        stack.pop()
    elif choice == "3":
        stack.find_top()
    elif choice == "4":
        stack.display_stack()
    elif choice == "5":
        print("Exiting program...")
        break
    else:
        print("Invalid choice, try again.")
```

```python
#implementing stack using deque

from collections import deque

class Stack:
    def __init__(self):
        self.stack = deque()

    def push(self, data):
        """Pushes an element onto the stack."""
        self.stack.append(data)

    def pop(self):
        """Removes and returns the top element of the stack."""
        if self.is_empty():
            print("Stack is empty")
            return None
        return self.stack.pop()

    def peek(self):
        """Returns the top element without removing it."""
        if self.is_empty():
            print("Stack is empty")
            return None
        return self.stack[-1]

    def is_empty(self):
        """Checks if the stack is empty."""
        return len(self.stack) == 0

    def size(self):
        """Returns the size of the stack."""
        return len(self.stack)

    def display(self):
        """Displays the stack."""
        print("Stack (top to bottom):", list(self.stack)[::-1])

# Example usage
stack = Stack()
stack.push(10)
stack.push(20)
stack.push(30)
stack.display()  # Output: Stack (top to bottom): [30, 20, 10]
print("Top element:", stack.peek())  # Output: 30
print("Popped element:", stack.pop())  # Output: 30
stack.display()  # Output: Stack (top to bottom): [20, 10]
```
