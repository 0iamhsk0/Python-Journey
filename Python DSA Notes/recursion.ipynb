```python
# Factorial of n natural numbers

class Solution:
    def factorial(self, n):
        """
        Function to calculate the factorial of n
        :param n: int, natural number
        :return: int, factorial of n
        """
        # Factorial of 0 or 1 is 1
        if n == 0 or n == 1:
            return 1

        # Calculate factorial using recursion
        return n * self.factorial(n - 1)
    

if __name__ == "__main__":
    # Create an instance of the Solution class
    sol = Solution()

    # Input from the user
    n = int(input("Enter a natural number to calculate its factorial: "))

    if n < 0:
        print("Factorial is not defined for negative numbers.")
    else:
        # Call the factorial function
        result = sol.factorial(n)
        print(f"The factorial of {n} is: {result}")
                                             
```

```python
# Reversing an array using iteration and recursion

class Recursion:
    # Iterative method
    def array_rev_iterative(self, arr):
        n = len(arr)
        start, end = 0, n - 1
        while start < end:
            # Swap the elements at the start and end
            arr[start], arr[end] = arr[end], arr[start]
            start += 1
            end -= 1
        return arr

    # Recursive method
    def array_rev_recursive(self, arr, start, end):
        if start >= end:
            return
        # Swap the elements
        arr[start], arr[end] = arr[end], arr[start]
        # Recursive call
        self.array_rev_recursive(arr, start + 1, end - 1)

# Creating an object of the class
rec = Recursion()

# Input array
arr = [1, 2, 3, 4, 5, 6]

# Reversing the array using the iterative method
reversed_arr_iterative = rec.array_rev_iterative(arr[:])  # Passing a copy of the array
print("Reversed array using iterative method:", reversed_arr_iterative)

# Reversing the array using the recursive method
arr_recursive = [1, 2, 3, 4, 5, 6]  # Resetting the array for recursion
rec.array_rev_recursive(arr_recursive, 0, len(arr_recursive) - 1)
print("Reversed array using recursive method:", arr_recursive)

# Using list comprehension to reverse the array
arr = [1, 2, 3, 4, 5, 6]
rev = [arr[i] for i in range(len(arr) - 1, -1, -1)]
print("Reversed array using list comprehension:", rev)

# Using a while loop to reverse the array
arr = [1, 2, 3, 4, 5, 6]
start, end = 0, len(arr) - 1
while start < end:
    arr[start], arr[end] = arr[end], arr[start]
    start += 1
    end -= 1
print("Reversed array using while loop:", arr)


```

### patterns in recursion:

1. Take not take. Eg: Print a subset
2. Count no. of ways
3. Backtracking

Let me break this down step by step and help you understand the Take-Not-Take pattern in recursion and how to translate it to Python. Additionally, I'll explain the concepts behind it and how the ans variable works in Python.

1. Understanding "Take-Not-Take" Recursion
This algorithm is used to generate all subsets (or powerset) of a given array. The idea is to traverse the array recursively, where:

At each step, you have two choices:

Take the current element (add it to the current subset).
Not take the current element (ignore it).
You keep doing this until you've considered all elements, at which point you add the current subset to the list of all subsets.

This concept is often referred to as "binary choice" because for each element, you can represent "take" as 1 and "not take" as 0. This is why your lecturer mentioned it's based on "bits"—it mimics the binary representation of subsets.

For example:

Input array: [1, 2, 3]
All subsets: [[], [1], [2], [3], [1, 2], [1, 3], [2, 3], [1, 2, 3]]
2. The ans Variable
In C++, ans = v<v<i>> is shorthand for declaring a 2D vector (list of lists) to store all subsets. In Python, you can represent this as a simple list of lists:

def generate_subsets(arr, idx, current, ans):
    """
    Generate all subsets of the array using recursion.
    :param arr: List[int], input array
    :param idx: int, current index in the array
    :param current: List[int], current subset being constructed
    :param ans: List[List[int]], list of all subsets
    """
    # Base condition: if we've processed all elements
    if idx == len(arr):
        ans.append(current[:])  # Append a copy of the current subset to the answer
        return

    # Take the current element
    current.append(arr[idx])
    generate_subsets(arr, idx + 1, current, ans)

    # Don't take the current element (backtrack)
    current.pop()
    generate_subsets(arr, idx + 1, current, ans)

# Driver code
arr = [1, 2, 3]
ans = []  # To store all subsets
generate_subsets(arr, 0, [], ans)
print("All subsets:", ans)
4. Explanation of the Code
Base Case:

If idx == len(arr), it means you've considered all elements in the array. At this point, you add the current subset (current) to ans.
Recursive Steps:

Take: Add arr[idx] to current and recurse with idx + 1.
Not Take: Remove the last added element (current.pop()) and recurse with idx + 1.
Backtracking:

After considering an element, you undo the addition (via pop) to explore other subsets. This ensures no duplicate subsets are generated.
5. How Is It Based on Bits?
The binary representation of subsets corresponds to the "take" and "not take" choices. Here's how:

For an array [1, 2, 3]:

Subsets can be represented as bits, where 1 means "take" and 0 means "not take":
000 → [] (empty subset)
001 → [3]
010 → [2]
011 → [2, 3]
100 → [1]
101 → [1, 3]
110 → [1, 2]
111 → [1, 2, 3]
Each recursive call essentially builds these combinations by deciding to include or exclude each element.

6. Dry Run
For arr = [1, 2]:

Start with current = [] and ans = [].
First recursive branch (Take):
Add 1 → Recurse → Add 2 → Base case reached → ans = [[1, 2]].
Backtrack → Remove 2 → Base case reached → ans = [[1, 2], [1]].
Second recursive branch (Don't Take):
Remove 1 → Add 2 → Base case reached → ans = [[1, 2], [1], [2]].
Backtrack → Remove 2 → Base case reached → ans = [[1, 2], [1], [2], []].
Output: [[1, 2], [1], [2], []].

7. Final Thoughts
The Take-Not-Take algorithm is a versatile way to generate subsets or solve problems like combinations, permutations, and more.
Using Python's list and append/pop makes it easy to work with subsets.
If you want further optimization or variations (e.g., with conditions), let me know!

```python
#take not take algo for printing all the subsequences/subsets:

def generate_subsets(arr, idx, temp, ans):
    """
    Generate all subsets of the array using recursion.
    :param arr: List[int], input array
    :param idx: int, temp index in the array
    :param temp: List[int], temp subset being constructed
    :param ans: List[List[int]], list of all subsets
    """
    # Base condition: if we've processed all elements
    if idx == len(arr):
        ans.append(temp[:])  # Append a copy of the temp subset to the answer
        return

    # Take the temp element
    temp.append(arr[idx])
    generate_subsets(arr, idx + 1, temp, ans)

    # Don't take the temp element (backtrack)
    temp.pop()
    generate_subsets(arr, idx + 1, temp, ans)

# Driver code
arr = [1, 2, 3]
ans = []  # To store all subsets
generate_subsets(arr, 0, [], ans)
print("All subsets:", ans)

```

```python
#print all the subets whose sum = k

def sum_subset(arr, temp, k, idx, current_sum):
    # Base Case: When index reaches the end of the array
    if idx == len(arr):
        if current_sum == k:
            ans.append(temp[:])  # Store a copy of the valid subsequence
        return

    # Take the current element (Include it in the subset)
    temp.append(arr[idx])
    sum_subset(arr, temp, k, idx + 1, current_sum + arr[idx])

    # Don't take the current element (Exclude it from the subset)
    temp.pop()  # Backtrack
    sum_subset(arr, temp, k, idx + 1, current_sum)

# Input Array
arr = [1, 2, 1, 2, 3, 2, 1]
k = 4
ans = []

# Function Call
sum_subset(arr, [], k, 0, 0)

# Print All Subsequences That Sum to k
print("Subsequences that sum to", k, "are:", ans)
```

### Count no. of ways:

base case:
1. if your cond matches return 1
2. else return 0

```python
#count number of subsequences present in the array with equal 0s and equal 1s

def count_seq(arr, idx, k):
    # Base Case: Reached end of array
    if idx == len(arr):
        return 1 if k == 0 else 0
    
    # Take the element
    if arr[idx] == 1:
        take = count_seq(arr, idx + 1, k + 1)  # Increase balance for 1
    else:
        take = count_seq(arr, idx + 1, k - 1)  # Decrease balance for 0
    
    # Don't take the element
    not_take = count_seq(arr, idx + 1, k)
    
    return take + not_take

# Input Array
arr = [1, 0, 1, 0, 1, 1]

# Function Call (Start from index 0 with k=0 balance)
print("Number of valid subsequences:", count_seq(arr, 0, 0))


```

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def generate_prime_composite(N):
    primes = []
    composites = []
    num = 2
    while len(primes) < (N // 2) + 1 or len(composites) < (N // 2) + 1:
        if is_prime(num):
            primes.append(num)
        else:
            composites.append(num)
        num += 1

    result = [0] * N
    prime_idx = 0
    composite_idx = 0

    for i in range(N):
        if i % 2 == 0:
            result[i] = primes[prime_idx]
            prime_idx += 1
        else:
            result[i] = composites[composite_idx]
            composite_idx += 1

    return result

# Number of spaces
N = 3

# Generate the sequence
sequence = generate_prime_composite(N)
print("Sequence with primes at even positions and composites at odd positions:", sequence)
def count_ways(sequence, idx, k):
    # Base Case: Reached end of sequence
    if idx == len(sequence):
        return 1 if k == 0 else 0
    
    # Take the element
    if sequence[idx] % 2 == 0:
        take = count_ways(sequence, idx + 1, k + 1)  # Increase balance for even
    else:
        take = count_ways(sequence, idx + 1, k - 1)  # Decrease balance for odd
    
    # Don't take the element
    not_take = count_ways(sequence, idx + 1, k)
    
    return take + not_take

# Function Call (Start from index 0 with k=0 balance)
print("Number of valid ways:", count_ways(sequence, 0, 0))
```
